```python

import csv
print("\n\n=== ASSIGNMENT 2 ===")


days = []
hours = []
consumptions = []


with open('consumption_data.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader) 
    
    for row in reader:
        day = int(row[0])      
        hour = int(row[1])      
        consumption = float(row[2])
        
        days.append(day)
        hours.append(hour)
        consumptions.append(consumption)

print("Data loaded successfully!")
print(f"Days: {days}")
print(f"Hours: {hours}")
print(f"Consumptions: {consumptions}")


hour_groups = {}


for i in range(len(hours)):
    current_hour = hours[i]
    current_consumption = consumptions[i]
    
    
    if current_hour not in hour_groups:
        hour_groups[current_hour] = []
    
    
    hour_groups[current_hour].append(current_consumption)

print("\n--- HOURLY GROUPING ---")
for hour in sorted(hour_groups.keys()):
    print(f"Hour {hour}: {hour_groups[hour]}")
    
    
    total_for_hour = 0
    for value in hour_groups[hour]:
        total_for_hour = total_for_hour + value
    
    average_for_hour = total_for_hour / len(hour_groups[hour])
    
    peak_for_hour = hour_groups[hour][0]
    for value in hour_groups[hour]:
        if value > peak_for_hour:
            peak_for_hour = value
    
    print(f"  Average: {average_for_hour:.2f}")
    print(f"  Peak: {peak_for_hour}")

day_groups = {}

for i in range(len(days)):
    current_day = days[i]
    current_consumption = consumptions[i]
    

    if current_day not in day_groups:
        day_groups[current_day] = []
    
    day_groups[current_day].append(current_consumption)

print("\n--- DAILY GROUPING ---")
for day in sorted(day_groups.keys()):
    print(f"Day {day}: {day_groups[day]}")
    
    
    total_for_day = 0
    for value in day_groups[day]:
        total_for_day = total_for_day + value
    
    
    average_for_day = total_for_day / len(day_groups[day])
    
    print(f"  Total: {total_for_day}")
    print(f"  Average: {average_for_day:.2f}")
```
