sum: sum() adds up numbers in a list.

i.	The power consumed by individual components is listed in a list. Find the total power consumed by adding readings.

```python

power_readings = [12, 5, 8, 10]
total_power = sum(power_readings)
print("Total Power Consumed:", total_power)
```

ii.	A list contains the values of a series of connected resistances. Find its equivalent resistance using the sum.

```python

resistances = [4, 6, 10] 
total_resistance = sum(resistances)
print("Equivalent Resistance:", total_resistance)
```
