. Lambda With map()
	Use map + lambda to transform a list of numbers.
	Use map + lambda to transform a list of strings.
	Use map + lambda on a list of tuples.

```python
nums = [1, 2, 3, 4]
result = list(map(lambda x: x * 2, nums))
print(result)
words = ["apple", "banana", "mango"]
result = list(map(lambda w: w.upper(), words))
print(result)
pairs = [(1, 2), (3, 4), (5, 6)]
result = list(map(lambda p: p[0] + p[1], pairs))
print(result)   # Output: [3, 7, 11]
```


4. Lambda With filter()
	Use filter + lambda to keep only even numbers.
	Use filter + lambda to keep only strings longer than 3 characters.
	Use filter + lambda on a mixed list of different types.

```python
nums = [1, 2, 3, 4, 5, 6]
result = list(filter(lambda x: x % 2 == 0, nums))
print(result)
words = ["cat", "lion", "sun", "apple"]
result = list(filter(lambda w: len(w) > 3, words))
print(result)
items = [1, "apple", 2.5, 8, "hi", 10]
result = list(filter(lambda x: type(x) == int, items))
print(result) 
```

5. Lambda With sorted() key
	Sort a list by length using lambda as key.
	Sort a list of tuples by their second element.
	Sort dictionary items by their values using lambda.

```python
words = ["apple", "hi", "banana", "cat"]
result = sorted(words, key=lambda w: len(w))
print(result)
pairs = [(1, 3), (4, 1), (2, 2)]
result = sorted(pairs, key=lambda x: x[1])
print(result)
data = {"a": 3, "b": 1, "c": 2}
result = sorted(data.items(), key=lambda x: x[1])
print(result)
```

9. Sort electrical components using lambda key
Sort this list of components by their watt rating using sorted() and a lambda key:
components = [("LED", 5), ("Fan", 60), ("Motor", 350), ("Bulb", 40)]
Then show the same sorting using operator.itemgetter.
State one advantage of itemgetter.

```python
components = [("LED", 5), ("Fan", 60), ("Motor", 350), ("Bulb", 40)]
sorted_by_watt_lambda = sorted(components, key=lambda x: x[1])
print("Sorted with lambda:", sorted_by_watt_lambda)
from operator import itemgetter
sorted_by_watt_itemgetter = sorted(components, key=itemgetter(1))
print("Sorted with itemgetter:", sorted_by_watt_itemgetter)
```

10. Convert sensor strings using map() + lambda
Convert a list of sensor voltage readings (strings) into floats and add +0.1 calibration offset:
readings = ["2.5", "5.0", "3.3"]
Show also the equivalent list comprehension.
Say which one looks cleaner.


```python
readings = ["2.5", "5.0", "3.3"]
```

```python
calibrated_readings_map = list(map(lambda x: float(x) + 0.1, readings))
print("Calibrated readings (map+lambda):", calibrated_readings_map)
calibrated_readings_list = [float(x) + 0.1 for x in readings]
print("Calibrated readings (list comprehension):", calibrated_readings_list)
```

11. Filter out low power values using filter() + lambda
From this list, keep only values above 50 W:
powers = [20, 75, 45, 120, 10]
Show the filter() version and a list-comprehension version


```python
powers = [20, 75, 45, 120, 10]
high_powers_filter = list(filter(lambda x: x > 50, powers))
print("High power values (filter+lambda):", high_powers_filter)
high_powers_list = [x for x in powers if x > 50]
print("High power values (list comprehension):", high_powers_list)
```

12. Use reduce() to combine values (total impedance calculation)
Using functools.reduce + lambda, compute the product of impedances in this list:
Z = [2, 3, 4]
Then show the simpler alternative using a loop or math.prod.
Write one short line on which is easier to understand.

```python
Z = [2, 3, 4]
```

```python
from functools import reduce
total_impedance_reduce = reduce(lambda x, y: x * y, Z)
print("Total impedance (reduce+lambda):", total_impedance_reduce)
total_impedance_loop = 1
for z in Z:
    total_impedance_loop *= z
print("Total impedance (loop):", total_impedance_loop)
import math
total_impedance_prod = math.prod(Z)
print("Total impedance (math.prod):", total_impedance_prod)
```
