Find Total Impedances
Add the three impedances.

```python
import cmath
Z1 = 10 + 5j     
Z2 = 4 - 3j       
Z3 = 7 + 2j    
Z_total = Z1 + Z2 + Z3
print("Total Impedance =", Z_total)
```

2:Print real part (resistive) and imaginary part (reactive). 

```python

print("Real Part (Resistive) =", Z_total.real)
print("Imaginary Part (Reactive) =", Z_total.imag)
```

3:Print magnitude and phase angle (use abs() and cmath.phase()).

```python

magnitude = abs(Z_total)
phase_angle = cmath.phase(Z_total)   

print("Magnitude =", magnitude)
print("Phase Angle (radians) =", phase_angle)
print("Phase Angle (degrees) =", phase_angle * 180 / 3.14159)
```
