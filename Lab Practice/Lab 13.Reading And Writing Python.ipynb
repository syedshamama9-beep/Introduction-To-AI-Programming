Solar Energy Generation Logging
A solar PV system records daily energy generation (kWh).

```python
import csv
from datetime import datetime, timedelta


load_file = "hourly_load_data.csv"

with open(load_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Timestamp", "Load_kW"])

start_time = datetime(2026, 1, 10, 0, 0)
with open(load_file, mode='a', newline='') as file:
    writer = csv.writer(file)
    for hour in range(24):
        timestamp = start_time + timedelta(hours=hour)
        load_value = 100 + hour * 5  
        writer.writerow([timestamp.strftime("%Y-%m-%d %H:%M"), load_value])
```

Task 7: Organize and Process CSV Data
Read the CSV file line by line and:
Count the total number of recorded days
Display each record with a serial number
The header must not be counted as a data record.

```python
energy_file = "solar_generation.csv"

with open(energy_file, mode='r') as file:
    reader = csv.reader(file)
    header = next(reader)  
    records = list(reader)

print(f"Total recorded days: {len(records)}\n")
for idx, row in enumerate(records, start=1):
    print(f"{idx}. Date: {row[0]}, Energy: {row[1]} kWh")
```

Task 8: Write Organized Data to a New CSV File
Create a new file named organized_energy_data.csv.
Write the processed and formatted energy records into this file in CSV format.
The original file must remain unchanged.

```python
organized_file = "organized_energy_data.csv"

with open(organized_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header) 
    for row in records:
        writer.writerow(row)
```
