task 4 parity checker

Digital systems often use parity bits to detect errors in data transmission. 
‚Ä¢ Even Parity ‚Üí If the number of 1s in a binary signal is even. 

```python

data = "101101"   
ones = data.count("1")
print("Binary Data:", data)
print("Number of 1s:", ones)
if ones % 2 == 0:
    print("Even Parity: TRUE (Number of 1s is even)")
else:
    print("Even Parity: FALSE (Number of 1s is odd)")
```

.Odd Parity ‚Üí If the number of 1s in a binary signal is odd.

```python

data = "101101"   
ones = data.count("1")
print("Binary Data:", data)
print("Number of 1s:", ones)
if ones % 2 == 1:
    print("Odd Parity: TRUE (Number of 1s is odd)")
else:
    print("Odd Parity: FALSE (Number of 1s is even)")
```

task no 9

Write a Python program that: 
:Reads a given directory path. 
:Checks all files in that directory. 
:Counts how many files are PDF files (ending with .pdf). 
:Prints the total number of PDF files.

```python
import os

directory = "D:\\"
pdf_count = 0

for root, dirs, files in os.walk(directory,topdown=True):
    try:
        for file in files:
            if file.lower().endswith(".pdf"):
                pdf_count += 1
    except PermissionError:
        continue
print(f"\nTotal PDF files in '{directory}' and its subfolders: {pdf_count}")
```
