2.	Ohm’s Law Calculator:
•	In your electrical lab, you need to calculate the voltage across a resistor.
•	Define a function that takes current (I) and resistance (R) and returns the voltage (V = I × R).
•	Call the function using positional, keyword, and default arguments.
•	Try mixing argument styles.
•	See what happens if you miss an argument

```python
def voltage(I, R=1):
    return I * R
# Positional
print(voltage(2, 5))
# Keyword
print(voltage(I=3, R=4))
# Default
print(voltage(5))
# Mixing
print(voltage(2, R=10))
```

Task 03:
A solar inverter monitors the battery’s state of charge (SOC).
•	Write a function charge_from_solar(SOC) that:
o	Prints "Charging" if SOC < 20%.
o	Prints "Battery full, no action" if SOC > 90%.
o	Prints "Battery charging, SOC = ---%" otherwise.
•	Add docstrings and function annotations.
•	Print the function’s docstring to check.

```python
def charge_from_solar(SOC: int) -> None:
    """
    Checks battery state of charge.
    """
    if SOC < 20:
        print("Charging")
    elif SOC > 90:
        print("Battery full, no action")
    else:
        print("Battery charging, SOC =", SOC, "%")
charge_from_solar(60)
print(charge_from_solar.__doc__)
```

Task 04:
	Resistance, Voltage, and Current Lists:
You are building a program to analyze five resistor values in a circuit.
•	Function 1: resistance():  return a list of 5 resistances.
•	Function 2: voltage():  return voltages across each resistor.
•	Function 3: current():  take both lists (Resistance & Voltage) and calculate currents.
•	Display all lists nicely in the main program.
•	Try two versions:
o	Return as a tuple.
o	Return as a dictionary with keys like "current_1".

```python
def resistance():
    return [10, 20, 30, 40, 50]
def voltage():
    return [5, 10, 15, 20, 25]
def current(R, V):
    return [V[i]/R[i] for i in range(5)]
R = resistance()
V = voltage()
I = current(R, V)
print("Resistance:", R)
print("Voltage:", V)
print("Current:", I)
```

Task 05:
In circuit design, you often calculate series resistance.
•	Write a function series() that takes a list or tuple of resistances.
•	Return the total resistance.
•	Add docstrings and annotations.
•	Print the function’s __annotations__.

```python
def series(res: list) -> float:
    return sum(res)
print(series([10, 20, 30]))
print(series.__annotations__)
```

Task 06:
	Parallel Resistance:
Now, calculate parallel resistance.
•	Write a function parallel() that takes a list or tuple of resistances.
•	Return the equivalent resistance.
•	Add docstrings and annotations.
•	Print the function’s __annotations__.

```python
def parallel(res: list) -> float:
    """Calculates parallel resistance"""
    total = 0
    for r in res:
        total += 1/r
    return 1/total
print(parallel([10, 20, 30]))
print(parallel.__annotations__)
```

Task 09:
	Solar Inverter Data (Tuple/List):
A solar inverter gives data as [SOC, solar_generation, price]. Example: [60, 2, 60].
•	Define a function that takes this list/tuple.

Task 10:
	Solar Inverter Data (Dictionary):
A better inverter sends data as:
{ "SOC": 60, "solar_generation": 2, "price": 60 }
•	Define a function that receives this dictionary.
•	Print the values in a clear format.
•	Use dictionary unpacking.
•	Add docstrings and annotations.
•	Print __annotations__.

```python
def inverter_dict(data: dict) -> None:
    print("SOC:", data["SOC"])
    print("Solar:", data["solar_generation"])
    print("Price:", data["price"])
inverter_dict({"SOC": 60, "solar_generation": 2, "price": 60})
print(inverter_dict.__annotations__)
```

Task 11:
	Solar Inverter Data (Keyword Arguments):
Sometimes, data is passed as:
solar(SOC=60, solar_generation=2, price=60)
•	Write a function that receives these keyword arguments as a dictionary.
•	Print the values in a nice format.
•	Add docstrings and annotations.
•	Print  annotations  

```python
def solar(**data: int) -> None:
    for key, value in data.items():
        print(key, ":", value)
solar(SOC=60, solar_generation=2, price=60)
print(solar.__annotations__)
```
