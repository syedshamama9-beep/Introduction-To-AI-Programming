Task 02:
Lambda With Multiple Arguments
Make a lambda that takes two inputs.
Make a lambda that takes three inputs.
Try using a keyword-only argument in a lambda.

```python

add = lambda x, y: x + y
print(add(3, 5))  

multiply = lambda x, y, z: x * y * z
print(multiply(2, 3, 4)) 

power = lambda x, *, exp=2: x ** exp
print(power(5))       
print(power(5, exp=3))  
```

Task 03:
Lambda With map()
Use map + lambda to transform a list of numbers.
Use map + lambda to transform a list of strings.
Use map + lambda on a list of tuples.

```python

nums = [1, 2, 3, 4]
print(list(map(lambda x: x**2, nums)))  

words = ["hello", "world"]
print(list(map(lambda s: s.upper(), words)))  

tuples = [(1, 2), (3, 4), (5, 6)]
print(list(map(lambda t: t[0]+t[1], tuples))) 
```

Task 04:
 Lambda With filter()
Use filter + lambda to keep only even numbers.
Use filter + lambda to keep only strings longer than 3 characters.
Use filter + lambda on a mixed list of different types

```python

nums = [1, 2, 3, 4, 5, 6]
print(list(filter(lambda x: x % 2 == 0, nums))) 

words = ["hi", "hello", "ok", "world"]
print(list(filter(lambda s: len(s) > 3, words)))  
mixed = [1, "a", 2, None, 3, "b"]
print(list(filter(lambda x: isinstance(x, int), mixed))) 
```

Task 05:
Lambda With sorted() key
Sort a list by length using lambda as key.
Sort a list of tuples by their second element.
Sort dictionary items by their values using lambda

```python

words = ["hi", "hello", "hey"]
print(sorted(words, key=lambda w: len(w)))  

tuples = [("a", 3), ("b", 1), ("c", 2)]
print(sorted(tuples, key=lambda t: t[1]))  

d = {"a": 3, "b": 1, "c": 2}
print(sorted(d.items(), key=lambda item: item[1]))  
```

Task 06:
Lambda With reduce()
Use reduce + lambda to combine all numbers in a list.
Use reduce + lambda to join multiple strings.
Use reduce + lambda to find a custom smallest or largest element.

```python
from functools import reduce

nums = [1, 2, 3, 4]
print(reduce(lambda x, y: x + y, nums))  

words = ["hello", "world"]
print(reduce(lambda x, y: x + " " + y, words))  

nums = [-10, 5, -20, 7]
print(reduce(lambda x, y: x if abs(x) > abs(y) else y, nums))  
```

Lab 08:
Sort electrical components using lambda key
Sort this list of components by their watt rating using sorted() and a lambda key:
components = [("LED", 5), ("Fan", 60), ("Motor", 350), ("Bulb", 40)]
Then show the same sorting using operator.itemgetter.
State one advantage of itemgetter.

```python
components = [("LED", 5), ("Fan", 60), ("Motor", 350), ("Bulb", 40)]

sorted_components = sorted(components, key=lambda c: c[1])
print(sorted_components)  

from operator import itemgetter
sorted_components2 = sorted(components, key=itemgetter(1))
print(sorted_components2)
```

Lab 09:
Convert sensor strings using map() + lambda
Convert a list of sensor voltage readings (strings) into floats and add +0.1 calibration offset:
readings = ["2.5", "5.0", "3.3"]
Show also the equivalent list comprehension.
Say which one looks cleaner.

```python
readings = ["2.5", "5.0", "3.3"]

converted = list(map(lambda x: float(x)+0.1, readings))
print(converted)  

converted2 = [float(x)+0.1 for x in readings]
print(converted2)
```

Lab 10:
Filter out low power values using filter() + lambda
From this list, keep only values above 50 W:
powers = [20, 75, 45, 120, 10]
Show the filter() version and a list-comprehension version.

```python
powers = [20, 75, 45, 120, 10]

high_powers = list(filter(lambda x: x > 50, powers))
print(high_powers)  

high_powers2 = [x for x in powers if x > 50]
print(high_powers2)
```

Lab 11:
 Use reduce() to combine values (total impedance calculation)
Using functools.reduce + lambda, compute the product of impedances in this list:
Z = [2, 3, 4]
Then show the simpler alternative using a loop or math.prod.
Write one short line on which is easier to understand.

```python
Z = [2, 3, 4]

total_impedance = reduce(lambda x, y: x*y, Z)
print(total_impedance) 

import math
total_impedance2 = math.prod(Z)
print(total_impedance2)
```
