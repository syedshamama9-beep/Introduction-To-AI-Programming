```python

print("=== ASSIGNMENT 1 ===")


import csv

consumption_list = []

with open('consumption_data.csv', 'r') as file:
    reader = csv.reader(file)
    
    
    next(reader)
    
    
    for row in reader:
    
        consumption_value = float(row[2])  
        consumption_list.append(consumption_value)

print("Consumption data loaded successfully!")
print(f"Total readings: {len(consumption_list)}")


total = 0
for value in consumption_list:
    total = total + value


average = total / len(consumption_list)


maximum = consumption_list[0]
for value in consumption_list:
    if value > maximum:
        maximum = value

minimum = consumption_list[0]
for value in consumption_list:
    if value < minimum:
        minimum = value

consumption_kwh = [value / 1000 for value in consumption_list]


print("\n--- RESULTS ---")
print(f"Total Consumption: {total} watts")
print(f"Average Consumption: {average:.2f} watts")
print(f"Maximum Consumption: {maximum} watts")
print(f"Minimum Consumption: {minimum} watts")
print(f"\nFirst 5 values in kWh: {consumption_kwh[:5]}")
```
